Note:- You don't have to return anything, you just have to print the array.

Example 1:

Input:
n = 0
Output:
already Zero
Example 2:

Input:
n = 4
Output:
3 2 1 0
Example 3:

Input:
n = -3
Output:
-3 -2 -1 0

```python
n = int(input())
if n==0:
    print("Already Zero")
elif(n>0):
    while n>=0:
        print(n)
        n-=1
else:
    while n<=0:
        print(n)
        n+=1
        
```

Q25>> You are given a list arr that contains integers. You need to return average of the non negative integers.

Examples:

Input: arr = [-12, 8, -7, 6, 12, -9, 14]
Output: avg = 10.0
Explanation: The non negative numbers are 8 6 12 14. The sum is 8+6+12+14 = 40, Average = 40/4 = 10.0

```python
arr1 = input().split()
sum = 0
count = 0
for i in range(len(arr1)):
    if int(arr1[i])>0:
        sum += int(arr1[i])
        count += 1
        
print(sum / count)
```
